```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import yaml

with open("../config.yml", "r") as file:
    config = yaml.safe_load(file)
    planck_const = config["params"]["hbar"]
    mass_elec = config["params"]["me"]
    well_length = config["params"]["Lp"]
    energy_level = config["params"]["energy_level"]
```

```python
# Wavefunction for a nth state
def wavefunction(n, x, L):
    """
    Computes the wavefunction for the nth energy level in a 1D infinite potential well.

    Params:
    n: Quantum number indexing the energy levels and wavefunctions particle.
    x: Position of the particle along the 1D well, [0,L]
    L: Length of the potential well

    Returns:
    Wavefunction psi
    """
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)
```

```python
# Probability density
def probability_density(psi):
    """
    Computes the probability density.

    Params:
    psi: Wavefunction

    Returns:
    Probability density
    """

    return psi**2
```

```python
# Define position array (x values)
x = np.linspace(0, well_length, 1000)  # 1000 points between 0 and L
```

```python
# Plot wavefunctions
plt.figure(figsize=(10, 6))

for n in range(1, energy_level+1):
    psi_n = wavefunction(n, x, well_length)
    plt.plot(x, psi_n, label=f"Wavefunction n={n}")

plt.title("Wavefunctions for a Particle in a 1D Infinite Potential Well")
plt.xlabel("Position (x)")
plt.ylabel("(x)")
plt.legend()
plt.grid()
plt.show()
```

```python

# Plot probability density
plt.figure(figsize=(10, 6))

for n in range(1, energy_level+1):
    psi_n = wavefunction(n, x, well_length)
    prob_density = probability_density(psi_n)
    
    plt.plot(x, prob_density, '--', label=f"Probability Density n={n}")

plt.title("Probability Densities for a Particle in a 1D Infinite Potential Well")
plt.xlabel("Position (x)")
plt.ylabel("|(x)|^2")
plt.legend()
plt.grid()
plt.show()
```

```python
def energylevel(n, L, m):
    return (n**2 * np.pi**2 * planck_const**2) / (2 * m * L**2)

for n in range(1, energy_level+1):
    print(f"Eigen-energies at E_{n} = {energylevel(n, well_length, mass_elec):.2e} J")
    
```
